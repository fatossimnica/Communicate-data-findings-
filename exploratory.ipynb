##### Improving column readability 
1. Converting the "df.duration_sec" to minutes
2. Rename "df.duration_sec" to "df.duration_min"
3. Rounding the numbers from the "duration_min" to two decimals.

```python
df.duration_sec= df.duration_sec / 60 # Converting the "df.duration_sec" to minutes
df.rename(columns={'duration_sec' : 'duration_min'},inplace=True) # Rename "df.duration_sec" to "df.duration_min"
df['duration_min']=df['duration_min'].apply(lambda x:round(x,2)) # Rounding the numbers from the "duration_min" to two decimals.
```

```python
# Average trip duration in minutes
df.duration_min.mean()
```

```python
plt.scatter(df.index, df['duration_min'])
plt.ylabel('Trip Duration in minutes')
plt.xlabel('Sample Count')
plt.suptitle('Trip duration')

plt.show()
                      
```

##### 2. Day of the trip 

We want to explore in what day are the most trips taken. In order to do that we need to extraxt the day from "start_time" variable.
