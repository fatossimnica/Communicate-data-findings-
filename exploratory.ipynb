### Data wrangling


##### Define
1. Converting the "df.duration_sec" to minutes
2. Rename "df.duration_sec" to "df.duration_min"
3. Rounding the numbers from the "duration_min" to two decimals.

###### Code

```python
df1.duration_sec= df.duration_sec / 60 # Converting the "df.duration_sec" to minutes
df1.rename(columns={'duration_sec' : 'duration_min'},inplace=True) # Rename "df.duration_sec" to "df.duration_min"
df1['duration_min']=df1['duration_min'].apply(lambda x:round(x,2)) # Rounding the numbers from the "duration_min" to two decimals.
```

##### 2. Day of the trip 

The first quastion we want to explore in what day are the most trips taken. In order to do that we need to extraxt the day from "start_time" variable.

In the next step to make the analyses more readable we converted the numbers that represent week days into name of the day.

```python
df1.start_time #overview of the column 
```

```python
df1['day_of_trip'] = pd.DatetimeIndex(df['start_time']).day # extracting the day from "start_time" variable.
```

In the step below we converted the numbers into weekdays names to the variable and later the anlyses more understandable.

```python

df1 = df1.replace({'day_of_trip': [1,8,15,22]}, 'monday')
df1 = df1.replace({'day_of_trip': [2,9,16,23]}, 'tuesday')
df1 = df1.replace({'day_of_trip': [3,10,17,24]}, 'wednesday')
df1 = df1.replace({'day_of_trip': [4,11,18,25]}, 'thursday')
df1 = df1.replace({'day_of_trip': [5,12,19,26]}, 'friday')
df1 = df1.replace({'day_of_trip': [6,13,20,27]}, 'saturday')
df1 = df1.replace({'day_of_trip': [7,14,21,28]}, 'sunday')

#1,8,15,22 = "monday"
#2,9,16,23 = "tuesday"
#3,10,17,24 = "wednesday"
#4,11,18,25 = "thursday"
#5,12,19,26 = "friday"
#6,13,20,27 = "saturday"
#7,14,21,28 = "sunday"
```
